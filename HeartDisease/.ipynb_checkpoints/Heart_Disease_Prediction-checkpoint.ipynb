```python
# importing libraries
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('/content/heart.csv')
```

```python
# print first 5 rows of the dataset
heart_data.head()
```

```python
# number of rows and columns in the dataset
heart_data.shape
```

```python
# getting some info about the data
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
# checking the distribution of Target Variable
heart_data['target'].value_counts()
```

1 --> Defective Heart

0 --> Healthy Heart

```python
plt.figure(figsize=(15,10))
sns.heatmap(heart_data.corr(), annot=True, cmap='Greens')
plt.show()
```

Splitting the Features and Target

```python
X = heart_data.drop(columns=['target','fbs','restecg','oldpeak','ca'], axis=1)
Y = heart_data['target']
```

Splitting the Data into Training data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
from sklearn.pipeline import Pipeline
# Create the pipeline with feature scaling and RandomForestClassifier
pipeline = Pipeline([
    ('classifier', RandomForestClassifier())
])

# Fit the model
pipeline.fit(X_train, Y_train)
```

Model Evaluation

```python
# accuracy on training data
X_train_prediction = pipeline.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print('Accuracy on Training data : ', training_data_accuracy)


# accuracy on test data
X_test_prediction = pipeline.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print('Accuracy on Test data : ', test_data_accuracy)
```

```python
# Saving model
pickle.dump(pipeline, open('Heart_Diseases_pipeline.pkl','wb'))
```

Building a Predictive System
