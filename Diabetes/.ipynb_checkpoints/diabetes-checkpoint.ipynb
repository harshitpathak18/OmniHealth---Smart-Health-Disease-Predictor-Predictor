## **Data Preprocessing**

```python
df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

```python
# Now, we can look at where are missing values
df.isnull().sum()
```

```python
# The missing values will be filled with the median values of each variable
def median_target(var):   
    temp = df[df[var].notnull()]
    temp = temp[[var, 'Outcome']].groupby(['Outcome'])[[var]].median().reset_index()
    return temp
```

```python
# The values to be given for incomplete observations are given the median value of people who are not sick and the median values of people who are sick.
columns = df.columns
columns = columns.drop("Outcome")
for i in columns:
    median_target(i)
    df.loc[(df['Outcome'] == 0 ) & (df[i].isnull()), i] = median_target(i)[i][0]
    df.loc[(df['Outcome'] == 1 ) & (df[i].isnull()), i] = median_target(i)[i][1]
```

```python
# Number of missing values
df.isnull().sum()
```

## Outlier Observation Analysis

```python
for feature in df:
    
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3-Q1
    lower = Q1- 1.5*IQR
    upper = Q3 + 1.5*IQR
    
    if df[(df[feature] > upper)].any(axis=None):
        print(feature,"yes")
    else:
        print(feature, "no")
```

```python
# Conducting a stand alone observation review for the Insulin variable
# Suppressing contradictory values
Q1 = df.Insulin.quantile(0.25)
Q3 = df.Insulin.quantile(0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
df.loc[df["Insulin"] > upper,"Insulin"] = upper
```

```python
# Determining the outliers between all variables with the LOF method
from sklearn.neighbors import LocalOutlierFactor
lof =LocalOutlierFactor(n_neighbors= 10)
lof.fit_predict(df)
```

```python
df_scores = lof.negative_outlier_factor_
np.sort(df_scores)[0:30]
```

```python
# Choosing the threshold value according to lof scores
threshold = np.sort(df_scores)[7]
threshold
```

## Feature Engineering

```python
df.head()
```

```python
y = df["Outcome"]
X = df.drop(columns = ["Outcome","BloodPressure"], axis = 1)
cols = X.columns
index = X.index
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, shuffle=True, test_size=0.2, stratify=y) 
```

```python
from sklearn.pipeline import Pipeline

# Define the pipeline steps
steps = [
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier())
]

# Create the pipeline object
pipeline = Pipeline(steps)

# Fit the pipeline to your training data (X_train, y_train)
pipeline.fit(X_train, y_train)

# Make predictions on new data (X_new)
y_pre = pipeline.predict(X_test)
```

```python
print("Testing Score",accuracy_score(y_test,y_pre))
print("Training Score",pipeline.score(X_train,y_train))
```
